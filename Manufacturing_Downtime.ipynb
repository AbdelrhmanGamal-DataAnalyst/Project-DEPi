```python
# Load necessary dataframes
# Assuming file_path is already defined in the notebook context.

df_lp = pd.read_excel(file_path, sheet_name="Line productivity")
df_line_dt = pd.read_excel(file_path, sheet_name="Line downtime")

# Clean column names for both dataframes (stripping whitespace)
df_lp.columns = df_lp.columns.str.strip()
df_line_dt.columns = df_line_dt.columns.str.strip()

# Calculate total downtime per batch from df_line_dt (long format)
# The 'Line downtime' sheet has 'Batch', 'Factor', 'total_min'.
# We need to sum 'total_min' for each 'Batch'.
df_batch_dt = df_line_dt.groupby('Batch', as_index=False)['total_min'].sum()
df_batch_dt = df_batch_dt.rename(columns={'total_min': 'downtime_min'})

# Prepare df_lp for merging
df_lp['Batch_num'] = pd.to_numeric(df_lp['Batch'], errors='coerce')

# Merge productivity data with batch downtime
df_merged = df_lp.merge(df_batch_dt[['Batch', 'downtime_min']],
                        left_on='Batch_num', right_on='Batch', how='left')

# Fill NaN downtime_min with 0 for batches that had no recorded downtime
df_merged['downtime_min'] = df_merged['downtime_min'].fillna(0)

# Ensure 'Date' column in df_merged is datetime type before extracting date
df_merged['Date_only'] = pd.to_datetime(df_merged['Date'], errors='coerce').dt.date

# Group by Date_only to get daily aggregates
df_day = df_merged.groupby('Date_only').agg(
    batches=('Batch_num','count'),
    downtime_min=('downtime_min','sum')
).reset_index()

# Analyze relationship between downtime and batches
x = df_day['downtime_min']
y = df_day['batches']

# Handle cases where there might not be enough variance in x to calculate slope
if len(x.unique()) > 1:
    slope = ((x - x.mean())*(y - y.mean())).sum() / ((x - x.mean())**2).sum()
    intercept = y.mean() - slope*x.mean()
else:
    # If all downtime values are the same, slope is effectively 0, predict mean batches
    slope = 0
    intercept = y.mean()

def predict_batches(predicted_downtime_min):
    return intercept + slope * predicted_downtime_min

# Example prediction
predicted_downtime_min = 200
print(f"Expected batches tomorrow: {predict_batches(predicted_downtime_min):.2f} if downtime = {predicted_downtime_min} minutes")
```

```python
#What percentage of downtime is caused by operator errors vs. non-operator errors? What are the top 3 causes in each category?

# Load necessary sheets
df_line_dt = pd.read_excel(file_path, sheet_name="Line downtime")
df_factors = pd.read_excel(file_path, sheet_name="Downtime factors")

# Clean column names for both dataframes (stripping whitespace)
df_line_dt.columns = df_line_dt.columns.str.strip()
df_factors.columns = df_factors.columns.str.strip()

# Ensure 'Factor' column in df_line_dt is numeric for merging
df_line_dt['Factor'] = pd.to_numeric(df_line_dt['Factor'], errors='coerce').astype('Int64')

# Use the actual "Operator Error" column for classification
df_factors["Category"] = df_factors["Operator Error"].apply(
    lambda x: "Operator error" if str(x).strip().lower() == "yes" else "Non-operator error"
)

# Merge factors with downtime data
df_merged = df_line_dt.merge(df_factors, on="Factor", how="left")

# Group by category
df_category = df_merged.groupby("Category")["total_min"].sum().reset_index()
df_category = df_category.rename(columns={'total_min': 'Downtime_minutes'})

# Calculate percentage
total_downtime = df_category["Downtime_minutes"].sum()
df_category["Percentage"] = (df_category["Downtime_minutes"] / total_downtime) * 100
df_category["Percentage"] = df_category["Percentage"].map(lambda x: f"{x:.2f}%")

# Find top 3 causes in each category
df_summary = df_merged.groupby(["Category", "Description"]).agg(
    Total_minutes=("total_min", "sum")
).reset_index()

top3_operator = df_summary[df_summary["Category"] == "Operator error"].nlargest(3, "Total_minutes")
top3_non_operator = df_summary[df_summary["Category"] == "Non-operator error"].nlargest(3, "Total_minutes")

# Display results
print("=== Downtime Percentage by Category ===")
print(df_category.to_string(index=False))

print("\n=== Top 3 Operator Error Causes ===")
print(top3_operator.to_string(index=False))

print("\n=== Top 3 Non-Operator Error Causes ===")
print(top3_non_operator.to_string(index=False))
```

```python
#Which products are associated with higher downtime rates than others?

# Load necessary sheets
# Use the file_path variable already defined in the notebook context
df_productivity = pd.read_excel(file_path, sheet_name="Line productivity")
df_line_downtime = pd.read_excel(file_path, sheet_name="Line downtime")

# Clean column names for consistency
df_productivity.columns = df_productivity.columns.str.strip().str.lower().str.replace(' ', '_')
df_line_downtime.columns = df_line_downtime.columns.str.strip().str.lower().str.replace(' ', '_')

# Calculate total downtime per batch from df_line_downtime
df_batch_downtime = df_line_downtime.groupby('batch', as_index=False)['total_min'].sum()
df_batch_downtime = df_batch_downtime.rename(columns={'total_min': 'downtime_minutes'})

# Merge productivity data with batch downtime
df_merged_products = df_productivity.merge(df_batch_downtime, on='batch', how='left')

# Fill NaN downtime_minutes with 0 for batches that had no recorded downtime
df_merged_products['downtime_minutes'] = df_merged_products['downtime_minutes'].fillna(0)

# The original code's approach: 'product' and 'downtime_min'
product_col = 'product'
downtime_col_for_analysis = 'downtime_minutes' # Using the calculated column

# Calculate min_events before dropping NaNs
# Ensure that product_col and downtime_col_for_analysis exist in df_merged_products
if product_col not in df_merged_products.columns or downtime_col_for_analysis not in df_merged_products.columns:
    raise ValueError(f"Missing required columns in merged dataframe: {product_col} or {downtime_col_for_analysis}")

min_events = df_merged_products.groupby(product_col)[downtime_col_for_analysis].count().min()
print(f"Minimum number of downtime events across products: {min_events}")

# Clean data (drop NaNs for relevant columns)
df_cleaned = df_merged_products.dropna(subset=[product_col, downtime_col_for_analysis])

# Calculate the mean based on the same number of events
def mean_fixed_events(group, n=min_events):
    # Ensure there are enough events in the group, otherwise take all available
    if len(group) < n:
        sample = group
    else:
        sample = group.head(n) # Takes the first n events as they appear after product sort
    return sample[downtime_col_for_analysis].mean()

fixed_avg = (
    df_cleaned.sort_values([product_col]) # Sort by product as in original, order within product not specified for head(n)
      .groupby(product_col, group_keys=False) # group_keys=False to avoid setting product as index
      .apply(mean_fixed_events)
      .reset_index()
)
fixed_avg = fixed_avg.rename(columns={fixed_avg.columns[-1]: 'avg_downtime_fixed_n'}) # Renaming the last column

# Sort results
fixed_avg = fixed_avg.sort_values('avg_downtime_fixed_n', ascending=False)
print("\nTable of averages after equalizing event count:\n")
display(fixed_avg)

# Plot Bar chart with values above bars
plt.figure(figsize=(10, 5))
bars = plt.bar(fixed_avg[product_col], fixed_avg['avg_downtime_fixed_n'], color=plt.cm.tab10.colors)

plt.xticks(rotation=45, ha='right')
plt.ylabel("Average Downtime (fixed n events)")
plt.title("Products Ranked by Equalized Average Downtime")

# Add values above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        f"{height:.1f}",
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

```python

df_productivity = pd.read_excel(file_path, sheet_name="Line productivity")
df_line_downtime = pd.read_excel(file_path, sheet_name="Line downtime")

df_productivity.columns = df_productivity.columns.str.strip().str.lower().str.replace(' ', '_')
df_line_downtime.columns = df_line_downtime.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert 'start_time' in df_productivity to datetime objects to extract hour
df_productivity['start_time_dt'] = pd.to_datetime(df_productivity['start_time'], format='%H:%M:%S', errors='coerce').dt.time
df_productivity['start_hour'] = pd.to_datetime(df_productivity['start_time'], format='%H:%M:%S', errors='coerce').dt.hour

# Define shifts based on start_hour of the *batch*
def get_shift(hour):
    if 8 <= hour < 20: # Example shifts: Morning (8-19), Night (20-7)
        return 'Morning (820)'
    else:
        return 'Night (208)'

df_productivity['shift'] = df_productivity['start_hour'].apply(get_shift)

# Merge df_line_downtime with df_productivity to get shift for each downtime event
df_merged_downtime_shifts = df_line_downtime.merge(
    df_productivity[['batch', 'shift']],
    on='batch',
    how='left'
)

# Clean data (drop NaNs for relevant columns, especially if a batch has no shift info)
df_merged_downtime_shifts = df_merged_downtime_shifts.dropna(subset=['total_min', 'shift'])

# Group by shift and calculate average downtime (total_min per downtime event)
shift_summary = (
    df_merged_downtime_shifts.groupby('shift')['total_min']
      .agg(['mean', 'std', 'count'])
      .reset_index()
      .rename(columns={'mean':'avg_downtime', 'std':'std_downtime', 'count':'n_events'})
      .sort_values('avg_downtime', ascending=False)
)

print("\n=== Average Downtime by 12-hour Shifts ===\n")
print(shift_summary.to_string(index=False, formatters={
    'avg_downtime': '{:.2f}'.format,
    'std_downtime': '{:.2f}'.format,
    'n_events': '{:.0f}'.format
}))

# Plot the comparison
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
bars = plt.bar(shift_summary['shift'], shift_summary['avg_downtime'],
               yerr=shift_summary['std_downtime'], capsize=6,
               color=['#4e79a7', '#f28e2b'])

plt.title("Average Downtime per Event by Shift (820 vs 208)")
plt.ylabel("Average Downtime (minutes per event)")
plt.xlabel("Shift")
plt.xticks(rotation=0)

# Add values above the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f"{height:.1f}",
             ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#What is the productivity of each operator (batches produced  working time)?
#Which operators have the highest throughput (total batches, total hours worked)?(Charlie)

# Load the 'Line productivity' sheet
df_productivity = pd.read_excel(file_path, sheet_name="Line productivity")
df_productivity.columns = df_productivity.columns.str.strip() # Clean column names

# Convert 'Date' to datetime.date object
df_productivity['Date'] = pd.to_datetime(df_productivity['Date']).dt.date

# Convert 'Start Time' and 'End Time' to time objects
df_productivity['Start Time'] = pd.to_datetime(df_productivity['Start Time'].astype(str), format="%H:%M:%S", errors="coerce").dt.time
df_productivity['End Time'] = pd.to_datetime(df_productivity['End Time'].astype(str), format="%H:%M:%S", errors="coerce").dt.time

# Drop rows where Start Time or End Time could not be parsed (i.e., are NaT)
df_productivity.dropna(subset=['Start Time', 'End Time'], inplace=True)

# Function to combine date and time for datetime calculation
def combine_dt(row, time_col):
    return pd.Timestamp.combine(row['Date'], row[time_col])

# Create full datetime objects for start and end
df_productivity['start_dt'] = df_productivity.apply(lambda r: combine_dt(r, 'Start Time'), axis=1)
df_productivity['end_dt'] = df_productivity.apply(lambda r: combine_dt(r, 'End Time'), axis=1)

# Adjust end_dt for overnight shifts
overnight = df_productivity['end_dt'] < df_productivity['start_dt']
if overnight.any():
    df_productivity.loc[overnight, 'end_dt'] += pd.Timedelta(days=1)

# Calculate duration in minutes
df_productivity['Total production time'] = (df_productivity['end_dt'] - df_productivity['start_dt']).dt.total_seconds() / 60.0

# Group by Operator and aggregate
dfg = df_productivity.groupby('Operator').agg(
    Batches_Produced=('Batch', 'count'),
    Total_Production_Time_minutes=('Total production time', 'sum')
).reset_index()

print("Operator Productivity and Throughput:")
display(dfg)
```
